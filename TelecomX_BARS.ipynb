#📌 Extracción

````python
from google.colab import drive
drive.mount('/content/drive')
#Iniciamos importando nuestra base de datos
import pandas as pd
import numpy as np
datos_iniciales=pd.read_json('/content/drive/MyDrive/ChallengeTelecomx/TelecomX_Data.json')
datos_iniciales.head(10)
````

#🔧 Transformación

````python
#Lo transformamos en un formato que pueda interpretar python para poder normalizarlo
normalización = datos_iniciales.to_dict('records')
````

````python
#Normalizamos nuestro documento:
datos_normalizados=pd.json_normalize(normalización)
datos_normalizados.head(10)
````

````python
#Consultamos los tipos de dato que se encuentran en el archivo:
datos_normalizados.info()
````

````python
columnas=list(datos_normalizados.columns)
columnas
````

````python
#Podemos ver que en la columna de "account.Charges.Total" tenemos tipo de dato "object"
# debemos cambiarlo a un flotante, primero debemos eliminar los datos nulos
datos_normalizados['account.Charges.Total']=datos_normalizados['account.Charges.Total'].replace(' ',np.nan)
datos_normalizados['account.Charges.Total']=datos_normalizados['account.Charges.Total'].astype(float)
````

````python
#Verificamos el cambio:
datos_normalizados.info()
````

````python
#finalmente, transformemos el resto de columnas a un formato que pueda leer python
#en este caso, serán categorias
cat_columns=['Churn', 'customer.gender', 'customer.Partner', 'customer.Dependents',
    'phone.PhoneService', 'phone.MultipleLines',
    'internet.InternetService', 'internet.OnlineSecurity', 'internet.OnlineBackup',
    'internet.DeviceProtection', 'internet.TechSupport',
    'internet.StreamingTV', 'internet.StreamingMovies',
    'account.Contract', 'account.PaperlessBilling', 'account.PaymentMethod']
for col in cat_columns:
  datos_normalizados[col]=datos_normalizados[col].astype('category')
````

````python
#comprobamos todos los cambios:
datos_normalizados.info()
````

````python
#Ahora cambiamos el nombre de las columnas por algo mas "legible"
datos_normalizados.to_csv('datos_normalizados.csv')
datos_normalizados.columns=["ID","Cancelacion","Genero_cliente","Old","Pareja","Dependientes",
                            "Tiempo_contrato","Sub_Tel","Sub_MultiTel","Sub_Wifi","Sub_Security","SubBackup",
                            "Sub_Proteccion","Sub_TechSup","Sub_Streaming","Sub_Mov","Cont_type",
                            "Online_Pay","Pay_meth","Pago_mensual","Pago_total"]
datos_normalizados.head(10)
````

````python
#Verifiquemos si existen datos nulos, así como la ubicación de ellos
datos_normalizados.isnull().sum()
````

````python
#Podemos ver que se encuentra dentro de "Pago_total", al ser una variable flotante y que indica
#el pago total, podemos sustituir estos datos nulos por 0.0
datos_normalizados['Pago_total']=datos_normalizados['Pago_total'].fillna(0.0)
datos_normalizados.isnull().sum()
````

````python
#Regresemos a nuestro formato JSON pero ya sin datos nulos y con nuestras columnas renombradas
datos_normalizados.to_json('/content/datos_normalizados_nonull.json',index=False)
````

````python
#Utilizamos el DF previamente creado y le asignamos a una variable
datos=pd.read_json('/content/datos_normalizados_nonull.json')
print(datos.isnull().sum())
````

````python
#ahora, agreguemos una columna de "Cuentas_Diarias" para entender el valor diario
#entre el pago mensual "Pago_mensual" y los dias al mes, tomando en cuenta 30
datos['Cuentas_Diarias']=datos['Pago_mensual']/30
datos.head(10)
````

````python
#Redondeemos nuestra nueva columna a 2 decimales
datos['Cuentas_Diarias']=datos['Cuentas_Diarias'].round(2)
datos.sample(5)
````

````python
#Podemos facilitar la lecutra cambiando los valores de "yes" y "no" por valores binarios (0 y 1)
#Primero almacenemos en una variable las columnas que tienen este tipo de valor
binarios=['Pareja','Dependientes','Sub_Tel','Sub_MultiTel','Sub_Wifi','Sub_Security','SubBackup',
          'Sub_Proteccion','Sub_TechSup','Sub_Streaming','Sub_Mov','Online_Pay']
#Ahora cambiemos el yes y el no por 0 y 1
datos[binarios] = datos[binarios].replace({'Yes': 1, 'No': 0})
datos['Cancelacion'] = datos['Cancelacion'].replace({'':'Situacion desconocida'})
datos.head(10)
````

````python
datos.sample(5)
````

````python
#Ahora transformemos todo lo demás para facilitar nuestra lectura:
datos['Genero_cliente']=datos['Genero_cliente'].replace({'Female': 'Femenino', 'Male': 'Masculino'})
datos['Cont_type']=datos['Cont_type'].replace({'Month-to-month': 'Mensual', 'One year': 'Anual', 'Two year': 'Bienal'})
datos['Pay_meth']=datos['Pay_meth'].replace({'Electronic check': 'Cheque Electronico', 'Mailed check': 'Cheque Por Correo',
                                             'Bank transfer (automatic)': 'Domiciliado Transferencia',
                                             'Credit card (automatic)': 'Domiciliado TDC'})
datos['Sub_Wifi']=datos['Sub_Wifi'].replace({"Fiber optic": 'Fibra Óptica', "No": 'Sin internet', "DSL": 'Telefónico'})
datos.sample(10)
````

````python
#Sigamos con la media
datos.median(numeric_only=True).round(2)
````

````python
#Finalmente con la mediana
datos.std(numeric_only=True).round(2)
````

````python
#Ahora veamos los datos que se encuentran en cada una de nuestras columnas importantes
print(datos['Genero_cliente'].value_counts())
print(datos['Sub_Wifi'].value_counts())
print(datos['Cont_type'].value_counts())
print(datos['Pay_meth'].value_counts())
````

````python
datos['Cancelacion'] = datos['Cancelacion'].replace({'No': 'Vigente', 'Yes': 'Cancelado'})
print(datos['Cancelacion'].value_counts())
````

````python
#Iniciemos viendo nuestra gráfica de cancelaciones
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
df=pd.DataFrame({'Cancelacion':['Vigente','Cancelado','Situacion desconocida'],'count':[5174,1869,224]})
fig, ax=plt.subplots(figsize=(8,4))
ax=sns.barplot(x='Cancelacion',y='count',data=df,palette='inferno')
for p in ax.patches:
    altura = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, altura + 40, f'{int(altura)}',
            ha='center', va='center', fontsize=6, fontweight='bold')
plt.title('Cancelaciones',fontsize=16)
plt.xlabel('Situación',fontsize=12)
plt.xticks(fontsize=9)
plt.ylabel('Cantidad',fontsize=12)
plt.yticks(fontsize=9)
ax.tick_params(axis='y',which='both',size=0)
fig.savefig('/content/grafico_cancelaciones.png',transparent=False, dpi=300,bbox_inches='tight')
plt.show()
````

````python
datos=datos.rename(columns={'Old':'Edad (años)','Pay_meth':'Metodo_pago','Cont_type':'Tipo_Contrato'})
datos['Edad (años)'] = datos['Edad (años)'].replace({1: 'Mayor de 65', 0: 'Menor de 65'})
datos['Pareja']=datos['Pareja'].replace({1:'Con pareja',0:'Solter@'})
datos['Dependientes']=datos['Dependientes'].replace({1:'Con dependientes',0:'Sin dependientes'})

columnas_categoricas=['Genero_cliente','Edad (años)','Pareja','Dependientes','Metodo_pago','Tipo_Contrato']
fig,axes=plt.subplots(2,3,figsize=(18,10))
axes=axes.flatten()
for i,col in enumerate(columnas_categoricas):
  sns.countplot(x=col,data=datos,palette='inferno',hue='Cancelacion',ax=axes[i])
  axes[i].set_title(f'Cancelacion por {col.capitalize()}',fontsize=12)
  axes[i].set_xlabel(col.capitalize(),fontsize=10)
  axes[i].set_ylabel('Cantidad',fontsize=10)
  axes[i].tick_params(axis='x',rotation=30,labelsize=10)
  axes[i].legend(title='Cancelacion',fontsize=9,title_fontsize=10)
plt.tight_layout()
fig.savefig('/content/grafico_categorias.png',transparent=False, dpi=300,bbox_inches='tight')
plt.show()
````

````python
#Ahora grafiquemos nuestros datos numéricos:
columnas_numericas=['Pago_total','Tiempo_contrato']
fig,axes=plt.subplots(1,2,figsize=(18,5))
axes=axes.flatten()
for i,col in enumerate(columnas_numericas):
  sns.boxplot(y=col,data=datos,palette='inferno',hue='Cancelacion',ax=axes[i],orient='v')
  axes[i].set_title(f'Cancelacion relacionada al {col.capitalize()}',fontsize=12)
  axes[0].set_ylabel('Pago total',fontsize=10)
  axes[1].set_ylabel('Tiempo contratado',fontsize=10)
  axes[i].tick_params(axis='x',rotation=30,labelsize=10)
  axes[i].legend(title='Cancelacion',fontsize=9,title_fontsize=10,loc='upper right')
plt.tight_layout()
fig.savefig('/content/grafico_numericos.png',transparent=False, dpi=300,bbox_inches='tight')
plt.show()
````

# Análisis de Churn de Clientes - TelecomX

## 📋 **INTRODUCCIÓN**

El presente análisis tiene como objetivo identificar los factores que influyen en la cancelación de servicios (churn) de los clientes de TelecomX. A través del análisis exploratorio de datos, buscamos comprender los patrones de comportamiento de los clientes que cancelan para desarrollar estrategias efectivas de retención.

---

## 🔧 **METODOLOGÍA**

### Herramientas Utilizadas

- **Fuente de datos:** API JSON desde repositorio GitHub
- **Ambiente:** Google Colab

### Proceso de Análisis
1. **Extracción:** Importación de datos desde API
2. **Transformación:** Limpieza y tratamiento de datos faltantes
3. **Carga:** Creación del DataFrame final para análisis
4. **Exploración:** Generación de visualizaciones y análisis estadísticos

---

## 📊 **ANÁLISIS EXPLORATORIO DE DATOS**

### Distribución General del Churn
- **Total de clientes:** 7,043
- **Clientes activos:** 5,174 (73.5%)
- **Clientes que cancelaron:** 1,869 (26.5%)
- **Situación desconocida:** 224 (3.2%)

**Insight clave:** Aproximadamente 1 de cada 4 clientes cancela el servicio, lo que representa una tasa de churn considerable que requiere atención inmediata.

### Análisis por Variables Demográficas

#### Género
- No existe diferencia significativa en las tasas de cancelación entre hombres y mujeres
- Ambos géneros mantienen patrones similares de retención

#### Edad
- **Clientes menores de 65 años:** Mayor propensión a cancelar
- **Clientes mayores de 65 años:** Menor tasa de cancelación
- **Interpretación:** Los adultos mayores tienden a ser más fieles a los servicios

#### Estado Civil y Dependientes
- **Clientes con pareja:** Menor tasa de cancelación
- **Clientes con dependientes:** Mayor retención del servicio
- **Clientes solteros sin dependientes:** Mayor riesgo de churn

### Análisis por Características del Servicio

#### Tipo de Contrato (Factor Más Crítico)
```
Contrato Mensual: ALTA tasa de cancelación
Contrato 12 meses: MEDIA tasa de cancelación  
Contrato 24 meses: BAJA tasa de cancelación
```

#### Método de Pago
```
Cheque Electrónico: ALTA cancelación
Cheque por Correo: MEDIA-ALTA cancelación
Domiciliado/Transferencia: BAJA cancelación
```

#### Tipo de Internet
- **Fibra Óptica:** Sorprendentemente presenta la mayor tasa de cancelación
- **DSL:** Menor tasa de cancelación
- **Sin internet:** Tasa moderada de cancelación

### Análisis Financiero

#### Pago Total vs Churn
- Los clientes que cancelan no necesariamente son los de menor gasto
- Existe variabilidad en los montos de pago entre clientes que permanecen y cancelan

#### Tiempo de Contrato
- **Clientes nuevos (0-12 meses):** Mayor riesgo de cancelación
- **Clientes establecidos (>24 meses):** Mayor fidelidad y retención

---

## 💡 **INSIGHTS RELEVANTES**

### 1. **Factor Contractual Determinante**
Los contratos mensuales son el principal predictor de churn. La flexibilidad que ofrecen también facilita la cancelación.

### 2. **Paradoja de la Fibra Óptica**
A pesar de ser un servicio premium, los clientes de fibra óptica cancelan más. Esto sugiere posibles problemas de calidad, expectativas no cumplidas o falta de soporte adecuado.

### 3. **Importancia de la Automatización**
Los métodos de pago automáticos están fuertemente correlacionados con mayor retención, sugiriendo que la comodidad influye en la permanencia.

### 4. **Perfil Familiar como Ancla**
Los clientes con responsabilidades familiares (pareja/dependientes) ven el servicio como más esencial y cancelan menos.

### 5. **Período Crítico de Retención**
Los primeros 12 meses son cruciales. Los clientes que superan este período tienden a mantenerse a largo plazo.

### 6. **Estabilidad en Adultos Mayores**
Los clientes >65 años representan un segmento de alta fidelidad, posiblemente por resistencia al cambio y valoración de la estabilidad.

---

## 🎯 **ESTRATEGIAS PARA REDUCIR EL CHURN**

### **Estrategia 1: Promoción de Contratos de Fidelización**
**Objetivo:** Migrar clientes de planes mensuales a anuales

**Acciones:**
- Ofrecer descuentos del 10-15% por contratos anuales
- Incluir beneficios adicionales (instalación gratuita, soporte prioritario)
- Campañas dirigidas a clientes con contratos mensuales y más de 6 meses de antigüedad

### **Estrategia 2: Mejora del Servicio de Fibra Óptica**
**Objetivo:** Reducir la alta cancelación en este segmento premium

**Acciones:**
- Implementar programa de seguimiento post-instalación
- Mejorar el soporte técnico especializado para fibra
- Realizar encuestas de satisfacción específicas para identificar puntos de dolor
- Considerar ajustes en expectativas durante la venta

### **Estrategia 3: Incentivos para Pagos Automáticos**
**Objetivo:** Migrar clientes a métodos de pago automatizados

**Acciones:**
- Descuentos del 3-5% por domiciliar pagos
- Simplificar el proceso de cambio de método de pago
- Campañas educativas sobre beneficios de automatización
- Recordatorios proactivos sobre vencimientos para pagos manuales

### **Estrategia 4: Programa de Retención Temprana**
**Objetivo:** Reducir churn en los primeros 12 meses

**Acciones:**
- Plan de bienvenida con seguimiento en meses 3, 6 y 9
- Ofertas especiales para nuevos clientes en período de prueba
- Soporte proactivo para resolver problemas iniciales
- Programa de referidos para nuevos clientes

### **Estrategia 5: Segmentación Familiar**
**Objetivo:** Aprovechar la estabilidad de clientes con familia

**Acciones:**
- Desarrollar planes familiares atractivos
- Ofertas especiales para parejas o familias
- Servicios adicionales orientados al hogar
- Programas de fidelidad que incluyan beneficios familiares

### **Estrategia 6: Retención de Adultos Mayores**
**Objetivo:** Mantener y expandir este segmento fiel

**Acciones:**
- Soporte especializado y personalizado
- Canales de atención preferencial
- Planes adaptados a necesidades específicas
- Programas de descuentos para adultos mayores

---

## 🏁 **CONCLUSIONES**

El análisis revela que el churn en TelecomX está principalmente impulsado por factores relacionados with la flexibilidad contractual, métodos de pago y calidad percibida del servicio.

**Los hallazgos más importantes son:**
- Los contratos mensuales representan el mayor riesgo de cancelación
- La fibra óptica requiere atención urgente por su alta tasa de churn
- Los pagos automáticos son un fuerte predictor de retención
- El perfil demográfico familiar es más estable

**La implementación de las estrategias propuestas debería:**
- Priorizar la migración a contratos anuales
- Resolver los problemas de satisfacción en fibra óptica
- Promover agresivamente los pagos automáticos
- Implementar un programa integral de retención temprana

Con estas acciones, TelecomX puede reducir significativamente su tasa de churn y mejorar la retención de clientes de manera sostenible.
